# Loading Dataset

```python
df=pd.read_csv('SampleSuperstore.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df.iplot(kind='bar',x='Ship Mode',y='Sales',xTitle="Ship Modes",yTitle="Sales in average",
         title="Sales vs Ship Mode",color=['red'],theme='solar')
```

## Observation
Highest sales are done through standard shipment and lowest were recorded with same day.

```python
plt.figure(figsize=(18,11))
sns.set_style('whitegrid')
sns.countplot(x='Region',data=df,hue='Segment')
```

### ***Moat sales were recorded in 'west'  Region in all threee sectors and most buys are of the consumers and least people have bought it for the home office***

```python
plt.figure(figsize=(22,12))
sns.set_style('whitegrid')
chart=sns.barplot(x='State',y='Sales',data=df)
chart.set_xticklabels(chart.get_xticklabels(),rotation=60)
```

### **Most sales are from the 'Wyomging' State while least are from South Dakota**

```python
plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
chart=sns.barplot(x='State',y='Profit',data=df,)
chart.set_xticklabels(chart.get_xticklabels(),rotation=60)
```

### Observation
***Most profit are coming from state of Vermont while least are coming from columbia***

```python
plt.figure(figsize=(15,8))
sns.set_style('whitegrid')
sns.barplot(x='Category',y='Sales',data=df,hue='Segment')
```
